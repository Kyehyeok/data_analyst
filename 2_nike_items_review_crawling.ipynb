```python
#Step 0. 필요한 모듈과 라이브러리를 로딩합니다.
import sys # 시스템
import os  # 시스템

import pandas as pd  # 판다스 : 데이터분석 라이브러리
import numpy as np   # 넘파이 : 숫자, 행렬 데이터 라이브러리

from bs4 import BeautifulSoup    # html 데이터를 전처리
from selenium import webdriver   # 웹 브라우저 자동화
import time    # 서버와 통신할 때 중간중간 시간 지연. 보통은 1초
from tqdm import tqdm_notebook   # for문 돌릴 때 진행상황을 %게이지로 알려준다.

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import chromedriver_autoinstaller
import subprocess
import shutil
```

```python
def new_driver():
    try:
        shutil.rmtree(r"c:\chrometemp")  #쿠키 / 캐쉬파일 삭제
    except FileNotFoundError:
        pass

    subprocess.Popen(r'C:\Program Files\Google\Chrome\Application\chrome.exe --remote-debugging-port=9222 --user-data-dir="C:\chrometemp"') # 디버거 크롬 구동


    option = Options()
    option.add_experimental_option("debuggerAddress", "127.0.0.1:9222")

    chrome_ver = chromedriver_autoinstaller.get_chrome_version().split('.')[0] #크롬 버전
    try:
        driver = webdriver.Chrome(f'./{chrome_ver}/chromedriver.exe', options=option)
        return driver
    except:
        chromedriver_autoinstaller.install(True)
        driver = webdriver.Chrome(f'./{chrome_ver}/chromedriver.exe', options=option)
        return driver
    
    driver.implicitly_wait(10)
    
    
```

### 에러 발생시 각 상황 따른 함수

```python
def find_error_1(now_url,error_url,driver):
    if error_url == driver.current_url:
        global open_driver
        driver.close()
        time.sleep(4)
        open_driver=new_driver()
        time.sleep(4)
        open_driver.get(now_url)
        time.sleep(4)
        return open_driver
    
    else:
        return driver
        

def find_error_2(now_url,error_url,driver):
    if error_url == driver.current_url:
        global open_driver
        
        driver.close()
        time.sleep(4)
        open_driver=new_driver()
        time.sleep(4)
        open_driver.get(now_url)
        time.sleep(4)
        review_arrow = open_driver.find_element_by_xpath("//h2[@data-click-name='review_learn more']")
        review_arrow.click()
        return open_driver
    else:
        return driver

def find_error_3(now_url,error_url,driver):
    if error_url == driver.current_url:
        global open_driver
        driver.close()
        time.sleep(4)
        open_driver=new_driver()
        time.sleep(4)
        open_driver.get(now_url)
        time.sleep(4)
        review_arrow = open_driver.find_element_by_xpath("//h2[@data-click-name='review_learn more']")
        review_arrow.click()
        time.sleep(4) 
        review_detail = open_driver.find_element_by_xpath('//*[@id="detail-review"]/div/div[2]/a')
        review_detail.click()
        time.sleep(4) 
        return open_driver
    else:
        return driver

def find_error_4(now_url,error_url,driver):
    
    if error_url == driver.current_url:
        global open_driver
        driver.close()
        time.sleep(4)
        open_driver=new_driver()
        open_driver.get(now_url)
        time.sleep(4)
        review_arrow = open_driver.find_element_by_xpath("//h2[@data-click-name='review_learn more']")
        review_arrow.click()
        time.sleep(4)
        review_detail = open_driver.find_element_by_xpath('//*[@id="detail-review"]/div/div[2]/a')
        review_detail.click()
        time.sleep(4)
        review_more = open_driver.find_element_by_css_selector("#load-more")
        review_more.click()
        time.sleep(4)
        return open_driver
    
    else:
        return driver
```

```python
items = pd.read_parquet("nike.parquet")

uni_items=items.drop_duplicates(['title','color','subtitle'],keep='last') 
#컬러가 달라도 제품의 디자인이 같으면 제품페이지 url이 다르지만 리뷰페이지가 하나로 통합되어 있다.
```

```python
driver= new_driver()
```

```python
items_review = []

error_url='https://www.nike.com/kr/error/no-access.html'
error_comment=['관련문의사항은 고객센터에 문의해주시면 친절히 안내해드리겠습니다.','고객상담전화 : 080-022-0182']

for url in uni_items.url:
    item_review_list = []
    star_review_list = []
    try:
        driver.get(url)
        driver=find_error_1(url,error_url,driver)
        #################################################
        time.sleep(4) 

        try:  ######응모 제품은  페이지에 화살표가 없음 
            ##멤버쉽 제품은 다르다 >>> 고객안내  or 리뷰
            review_arrow = driver.find_element_by_xpath("//h2[@data-click-name='review_learn more']")
            review_arrow.click()# 리뷰 화살표 커서 클릭        

            driver=find_error_2(url,error_url,driver)
            #################################################
            time.sleep(4)

            try:# 더 많은 리뷰 있는지 없는지 (없으면 리뷰가 1도 없음.)

                review_detail = driver.find_element_by_xpath('//*[@id="detail-review"]/div/div[2]/a')
                review_detail.click()  # 더 많은 리뷰 보기 클릭

                driver=find_error_3(url,error_url,driver)
                #################################################
                time.sleep(4)

                while True:
                    try: #리뷰가 많으면 더보기 버튼이 있기 때문에 클릭하여 아래의 댓글도 확인해야함
                        review_more = driver.find_element_by_css_selector("#load-more")
                        review_more.click() # 더보기 버튼 클릭 

                        driver=find_error_4(url,error_url,driver)
                        time.sleep(4)

                        service_error_comment = driver.find_element_by_css_selector('#review-list > li:last-child').text #### li:last-child >div 등등 ㅆ

                        if service_error_comment in error_comment:
                            review_list = driver.find_elements_by_css_selector("#review-list > li > div > div.read-list.uk-width-large-2-3 > p.read-comment")
                            star_list = driver.find_elements_by_css_selector("#review-list > li > div > div.read-list.uk-width-large-2-3 > dl > dt.review-filter.star.like > i.brz-icon-star_small.star-5.per")

                            
                            
                            driver.close()  #quit 과 close의 차이
                            time.sleep(4)
                            driver= new_driver()
                            break

                    except:  # 하단에 더보기 클릭 없으면 태그 뽑기
                        #리뷰 리스트
                        review_list = driver.find_elements_by_css_selector("#review-list > li > div > div.read-list.uk-width-large-2-3 > p.read-comment")
                        star_list = driver.find_elements_by_css_selector("#review-list > li > div > div.read-list.uk-width-large-2-3 > dl > dt.review-filter.star.like > i.brz-icon-star_small.star-5.per")

                        break

                for review ,star in zip(review_list,star_list): #뽑은 태그의 text 모두 수집
                    item_review_list.append(review.text)
                    star_review_list.append(star.get_attribute('style'))

            except :
                item_review_list.append('리뷰가 없습니다')
                star_review_list.append('')
        except:
            item_review_list.append('응모 제품입니다')
            star_review_list.append('')
    except:

        item_review_list.append('url 소멸')
        star_review_list.append('')

    items_review.append({
                        'url' : url,
                        'review': item_review_list,
                        'star' : star_review_list
                })
```
